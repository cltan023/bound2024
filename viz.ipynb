```python
upper_bounds = []
for i in range(1, 301):
    diam_a = (train_loss_vec_hist[0] - train_loss_vec_hist[i]).norm(p=2).item()
    lip = max(stochastic_grad_norm_hist[:i]) * np.sqrt(1024.0)
    # dim_w = dimensions[i-1]
    dim_w = 3.5
    
    alpha = np.log2(1.0 / diam_a * len(train_loss_vec_hist[0]) ** 0.5 * lip)
    alpha = alpha ** 0.5
    beta = diam_a * 12 / len(train_loss_vec_hist[0])
    
    upper_bound = (2.0 * dim_w) ** 0.5 * beta * (alpha + 1.0 / alpha)

    upper_bounds.append(upper_bound)

predicted_loss = train_loss + torch.tensor(upper_bounds)

predicted_acc = []
for i in range(len(predicted_loss)):
    idx = np.argmin(np.abs(train_loss - predicted_loss[i]))
    predicted_acc.append(metrics[1][idx])

print(f'validation accuracy: {metrics[0][-1]:.2f}% and predicted accuracy: {predicted_acc[-1]:.2f}%')

plot_data.append([metrics[1], metrics[0], predicted_acc[:len(metrics[0])]])

# fig, ax = plt.subplots(1, 1, figsize=(4, 3))
# ax.plot(metrics[1], '--', color='r', label='training set')
# ax.plot(metrics[0], '-.', color='b', label='validation set')
# ax.plot(predicted_acc[:len(metrics[0])], ':', color='cyan', label='prediction')

# ax.legend(frameon=False)
# ax.set_xlabel('Epoch')
# ax.set_ylabel('Accuracy (%)')

# plt.tight_layout()

# if not os.path.exists('figs'):
#     os.mkdir('figs')
# plt.savefig(os.path.join('figs','vit_imagenet.pdf'), format='pdf', bbox_inches='tight')
```
